```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# seaborn is advance metplot library
# it is built on top of matplotlib
# it provides a high-level interface for drawing attractive statistical graphics
sns.get_dataset_names()
```

```python
df_titanic=sns.load_dataset('titanic')
df_titanic.head()
```

```python
#sns.countplot(data=df_titanic, x="sex") #hue parameter is used to add another categorical variable
sns.countplot(data=df_titanic, x="sex", hue='alive', palette=sns.color_palette("mako", 2))
plt.title('Titanic Survival analysis')
plt.show()
```

```python
# sns.barplot(data=df_titanic,x='class',y='fare',hue='alive',palette=sns.color_palette("mako", 2))
sns.barplot(data=df_titanic,x='class',y='fare',hue='alive',palette=sns.color_palette("mako", 2),errorbar=('ci',0)) #this will remove the error bars
# # ci = confidence interval used to show the uncertainty of the estimate and show the fluctuation in the data
# # by default it is 95% confidence interval
plt.grid('r',alpha=0.3) #this will add grid lines to the plot
```

```python
plt.Figure(figsize=(20, 10))
sns.violinplot(data=df_titanic,x='class',y='fare',hue='sex',orient='v')
# plt.xticks(0,450,10)
plt.grid('r',alpha=0.3)
plt.show()
```

```python
df_pen=sns.load_dataset('penguins')
df_pen.head()
```
