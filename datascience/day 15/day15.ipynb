```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import *
```

```python
data = {'Study_hrs':range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}

df = pd.DataFrame(data)
df
```

```python
# ML model build?: x, y: +ve, -ve
# Scatter Plot
plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker = 'o', color = 'r')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

best fit line is those lines whose margine or error is low 

```python
# mae mean absolute error 
# mse mean square error 
# mse is comman use for calculate pentalty 
# rmse is a upgraded version of mse here we just add a square root in the total error 

# equation of line -> y=mx+c we also say w in ML w= weight 
# m= slope of line 
# c= x intercept of y mean what the value of y when x is 0
# x,y are coordinate

# simple regression model is same as equation of time but here we say c to betanot and m= beta1 
# y=beta_not + beta_1 X
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
X=df[["Study_hrs"]]
y= df["Marks Obtained"]
model.fit(X,y)
```

```python
X=[[5.5]]
m=model.coef_
c=model.intercept_
print(m,c)
```

```python
std=[[5.2],[5.6]]
mark=m*std+c
print(mark)
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours: ')

std_hr = [[float(study_hrs)]]

ans = model.predict(std_hr)
marks = round(ans[0],2)

if marks > 100:
  marks = 100

elif marks<0:
  marks = 0

else:
  marks = marks

print(f'Your Predicted marks is: {marks}')
```
