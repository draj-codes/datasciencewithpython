```python
from googlesearch import search
title = input("Enter Search title: ")
data= search(title,num_results=10)
all_data=[]
for i in data:
    all_data.append(i+'\n')

# mode x is used to create only file if file exist then it show error if not exist then create a file
f= open(f'{title}.txt','w')
f.writelines(all_data)
f.close()
# print (all_data)
print("Data done")
```

```python
url='''https://picsum.photos/1080/720'''
import requests as rq
n=5
for i in range(n):
    ig_data=rq.get(url).content
    f=open(f'img_{i}.jpg','wb')
    f.write(ig_data)
    f.close()
print('done')
```

```python
url_path = '''https://www.python.org/ftp/python/3.13.5/python-3.13.5-amd64.exe'''

# context manager
import requests as r
with open('dev.exe','wb') as file:
    file.write(r.get(url_path).content)

print('Done')
```
