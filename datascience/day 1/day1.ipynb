```python
a=1234 #int
print(a,type(a))

b=123.3 #float
print(b,type(b))

c=1245+2j #complex
print(c,type(c))

d='Hello python' #string
print(d,type(d))
```

```python
#tuple
a=("Hello",321,12.3)
print(a,type(a))
#typecast
a=list(a)
print(a,type(a))
```
